```python
import matplotlib.pyplot as plt #3
import contextily as ctx #4
import geopandas #5
```

```python
#fonte dos dados https://dados.gov.br/dataset/pivos-mapeados6
df=geopandas.read_file("aaa.shp") #6
```

```python
df = df.to_crs(epsg=3857)#7
```

```python
ax = df[df['NM_ESTADO']=='DISTRITO FEDERAL'].plot(figsize=(10,15), alpha = 0.5, edgecolor = 'k') #8
```

```python
ctx.add_basemap(ax, zoom = 12) #9
plt.savefig('MapaDF.png')
```
